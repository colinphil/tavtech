# Pandas Interface
One of the coolest features in seaborn is the fact that we can pass a dataframe and then choose the columns to be our x, y values. We will now explore this with some of the more common plots out there.

```python
df = pd.DataFrame({
    'quarter' : [0, 1, 2, 0, 1, 2],
    'earnings' : [2, 85, 325, 80, 47, 1],
    'company' : ['G', 'G', 'G', 'B', 'B', 'B']
})

df.head()
```

## Lineplot
We first take a look at the `seaborn` line plot which is very good for trend searching. It is very useful when trying to analyze time series data which unfortunately we won't have time to go into.  
Notice how the `hue` parameters acts as a groupby mechanism.

```python
sns.lineplot(x='quarter', y='earnings', hue='company', data=df)
```

## Scatter plots

```python
df = pd.DataFrame({
    'x' : [0, 1, 2, 3, 4, 5],
    'y' : [11, 14, 17, 19, 20, 21],
    'size' : [100, 200, 300, 400, 500, 600],
    'color' : [10, 10, 20, 20, 20, 20]
})

df.head()
```

In the scatter plot we can choose a column which serves as the size of each point.

```python
sns.scatterplot(x='x', y='y', size='size', data=df)
```

And we can also set a color for each point.

```python
sns.scatterplot(x='x', y='y', size='size', hue='color', data=df)
```

## Bar plots
Very similar to what we just saw, we can easily create informative bar graph using seaborn

## distplot 
The seaborn distplot method combines matplotlib hist function, with a smart binning pick and a kde fit. But it works on series rather then DataFrames.

```python
values = np.random.randn(1000) * 10
sns.distplot(values)
```

But, If we want to get fancier we can use seaborn FacetGrid. We will not dive into this topic, what the facet grid does is grouping 

```python
# Load the tips dataset.
tips = sns.load_dataset("tips")

# Create a fcaet grid, creating the following groups:
g = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
bins = np.linspace(0, 60, 13)

# Show how the total_bill values behave in each group.
g.map(sns.distplot, "total_bill", bins=bins)
```

## Pair-plot
A great function for quickly visualizing a classification dataset is by using seaborn pair plot. This plot will gave a scatter plot of every possible combination of numeric variable along side each variable "distribution".

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
sns.pairplot(iris, hue='species')
```

So what did we get back?  
A scatter plot of each 2 variables, and the diagonal is the distribution of that variable under each class
