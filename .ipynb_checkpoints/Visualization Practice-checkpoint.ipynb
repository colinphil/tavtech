# Example of Visualizing Data
 Generate synthetic data for visualization practice

```python
import numpy as np
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load nba statistics as csv and create dataframe
data = pd.read_csv('/home/colinphillips17/Downloads/sportsref_download.csv', header = 1)
data.head()
```

```python
#sort data from earliest to latest
data = data.sort_values("Season",ascending = True)
```

```python
#create line graph of fg% across time
data["Season"] = data["Season"].str.slice(0,-3)
data["Season"] = pd.to_numeric(data["Season"],errors="coerce")
```

```python
seaborn.lineplot(x="Season",y="FG%",data = data)
```

```python
seaborn.lmplot(
    x='Season', 
    y='Age', 
    data=data, 
    ci=False, 
);
```

```python
seaborn.scatterplot(x="FGA",y="FG%",data = data)
```

```python
seaborn.scatterplot(x="Wt",y="3P%",data=data)
```

```python
fig, axs = plt.subplots(ncols=3,figsize = (20,10))
seaborn.lineplot(x='Season', y='Age', data=data, ax=axs[0])
seaborn.lineplot(x='Season', y='Pace', data=data, ax=axs[1])
seaborn.lineplot(x='Season',y='PTS', data=data, ax=axs[2])
axs
```

```python
seaborn.pairplot(data)
```
